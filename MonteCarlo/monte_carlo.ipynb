```python
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
SIZE_OF_SAMPLE  = 10000

planning = (10, 2)
design = (24,2)
dev = (70, 5)
testing = (24, 2)
deployment = (7, 2)


# Right out the different types of Sets for each phase
# Then create a list of sims based on time.
tasks = [planning, design, dev, testing, deployment]


sim_list = [[] for x in range(SIZE_OF_SAMPLE)]
for i in range(SIZE_OF_SAMPLE):
    for task in tasks:
        sim_list[i].append(np.random.normal(task[0], task[1]))
    sim_list[i] = sum(sim_list[i])


print(sim_list)
```

```python
# Creates the dataframe

project_df = pd.DataFrame({'outputs': sim_list})
```

```python


fig = px.histogram(project_df, x="outputs", nbins=50,
                   title="Simulation Project Completion times",
                   labels={"outputs":"Project Completion Time (Days)"} 
                   )


# Update layout to add the y-axis label
fig.update_layout(
    yaxis_title="Frequency of Simulation"
)


fig.show()
```

```python
# Box Plot

boxplot = px.box(project_df,
                points="all",
                 y="outputs",
                 labels={"outputs": "Project Completion Time (Days)"},
                 title = "Simulation Project Completion times",
                 )

boxplot.show()
```
